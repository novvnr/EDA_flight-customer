```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
# Memuat preprocessed data
data = pd.read_csv("preprocessed_customer_booking.csv", encoding='latin1')
data.head()
```

```python
# Membuat heatmap korelasi tiap kolom pada data
plt.figure(figsize=(12, 10))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Heatmap Korelasi Fitur')
plt.show()
```

```python
numeric_columns = ['num_passengers', 'purchase_lead', 'length_of_stay', 'flight_hour', 'flight_duration']
cat_columns = ['sales_channel', 'trip_type', 'route', 'booking_origin', 'wants_extra_baggage', 'wants_preferred_seat', 'wants_in_flight_meals', 'booking_complete']

```

```python
# Distribusi variabel numerik
plt.figure(figsize=(15, 10))
for i, column in enumerate(numeric_columns, 1):
    plt.subplot(2, 3, i)
    sns.histplot(data[column], kde=True)
    plt.title(f'Distribusi {column}')
plt.tight_layout()
plt.show()
```

```python
# Visualize pairplot for selected features
key_features = ['num_passengers', 'purchase_lead', 'length_of_stay', 'flight_hour', 'flight_duration']
sns.pairplot(data[key_features], diag_kind='kde', plot_kws={'alpha': 0.6, 's': 80, 'edgecolor': 'k'})
plt.suptitle('Pairplot of Key Features')
plt.show()
```

```python
# Plot countplots for categorical features
plt.figure(figsize=(15, 20))
for i, col in enumerate(cat_columns, start=1):
    plt.subplot(4, 3, i)
    sns.countplot(x=data[col])
    plt.title(f'Countplot of {col}')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Step 6: Bivariat Analysis

# Scatter plots for numerical features against the target variable
plt.figure(figsize=(15, 10))
for i, col in enumerate(numeric_columns, start=1):
    plt.subplot(3, 2, i)
    sns.scatterplot(x=data[col], y=data['booking_complete'])
    plt.title(f'{col} vs Booking Complete')
plt.tight_layout()
plt.show()
```

```python
# Distribusi variabel numerik
plt.figure(figsize=(15, 10))
for i, column in enumerate(numeric_columns, 1):
    plt.subplot(2, 3, i)
    sns.histplot(data[column], kde=True)
    plt.title(f'Distribusi {column}')
plt.tight_layout()
plt.show()

```
